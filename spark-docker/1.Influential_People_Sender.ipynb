```python
class TweetsListener(StreamListener):
  def __init__(self, csocket):
      self.client_socket = csocket
  def on_data(self, data):
      try:
          msg = json.loads( data )  
          user_name=str(msg['user']['screen_name']) # Get the user name 
          follower_count= msg['user']['followers_count']  # Get the follower count
          influential_people=[] 
          influential_people.append(user_name)  # Append user_name to the list 
          influential_people.append(follower_count) # Append user_name to the list 
          user_follower=influential_people[0]+","+str(influential_people[1])+"\n"  # Concatenate user_name "," and follower_count
          print(user_follower)
            
          data=str.encode(user_follower) # Convert the string to bytes
          self.client_socket.send(data) # Send to Spark Streaming 
          return True
        
      except BaseException as e:
          print("Error on_data: %s" % str(e))
      return True
  def on_error(self, status):
      print(status)
      return True
```

```python
def sendData(c_socket):
  auth = OAuthHandler(consumer_key, consumer_secret)
  auth.set_access_token(access_token, access_secret)
  twitter_stream = Stream(auth, TweetsListener(c_socket))
  #twitter_stream.filter(track=['modi'])
  twitter_stream.sample()  # Get all the tweets 
```
